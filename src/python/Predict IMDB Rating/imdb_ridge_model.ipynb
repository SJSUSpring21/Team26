```python
import pandas as pd
import numpy as np
```

```python
data_movies=pd.read_csv("unique10000.csv")
```

```python
data_movies.tail()
```

```python
data_movies['Budget'] = data_movies['Budget'].replace(0,data_movies['Budget'].mean())
data_movies['Revenue_US'] = data_movies['Revenue_US'].replace(0,data_movies['Revenue_US'].mean())
data_movies['Revenue_World'] = data_movies['Revenue_World'].replace(0,data_movies['Revenue_World'].mean())


```

```python
data_movies["Movie"]= data_movies['Movie'].astype('category')
data_movies["Movie_cat"]= data_movies["Movie"].cat.codes

data_movies["Genre"]= data_movies['Genre'].astype('category')
data_movies["lGenre_cat"]= data_movies["Genre"].cat.codes

data_movies["Director"]= data_movies['Director'].astype('category')
data_movies["Director_cat"]= data_movies["Director"].cat.codes

data_movies["Actor1"]= data_movies['Actor1'].astype('category')
data_movies["Actor1_cat"]= data_movies["Actor1"].cat.codes

data_movies["Actor2"]= data_movies['Actor2'].astype('category')
data_movies["Actor2_cat"]= data_movies["Actor2"].cat.codes
```

```python
data_movies.head()
```

```python
X= data_movies[["Movie_cat","lGenre_cat","Director_cat","Actor1_cat","Actor2_cat","Budget","Revenue_US","Revenue_World"]]
y= data_movies["Rating"]
```

```python
from sklearn.model_selection import train_test_split

# Splitting the training data into train set and validation set
x_train, x_val, y_train, y_val = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
from sklearn import linear_model
ridge = linear_model.Ridge(alpha=.5)
```

```python
ridge=ridge.fit(x_train,y_train)
```

```python
y_pred_ridge=ridge.predict(x_val)
```

```python
import pickle
filename = 'finalized_model.sav'
pickle.dump(ridge, open(filename, 'wb'))

loaded_model = pickle.load(open(filename, 'rb'))
```

```python
def stringToInt(string):
    integer = 0
    #try seeing if string value given is already a number if so the output would be
    try:
        #means string value given is already a int
        integer = int(string)
    except:
        string = string.lower()
        for i in string:
            integer += ord(i)
    return integer
```

```python
Movie = input("Enter the name of the Movie: ")
Movie = Movie.lower()
Movie = stringToInt(Movie)

Genre = input("Enter the name of the Genre: ")
Genre = Genre.lower()
Genre = stringToInt(Genre)

Director = input("Enter the name of the Director: ")
Director = Director.lower()
Director = stringToInt(Director)

Actor1 = input("Enter the name of the Actor1: ")
Actor1 = Actor1.lower()
Actor1 = stringToInt(Actor1)

Actor2 = input("Enter the name of the Actor2: ")
Actor2 = Actor2.lower()
Actor2 = stringToInt(Actor2)


Budget = int(input("Enter the budget for the movie: "))
Revenue_US = int(input("Enter the Revenue_US for the movie: "))
Revenue_World = int(input("Enter the Revenue_World for the movie: "))


prediction = ridge.predict([[Movie,Genre,Director,Actor1,Actor2,Budget,Revenue_US,Revenue_World]])

print(prediction[0]/10 * 10)
```
